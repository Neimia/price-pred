```python
import pandas as pd #это импорты
import numpy as np # если что код сделан по pep8 поэтому тут такие странные отступы
import torch # чтобы сделать по этому стандарту я в pycharm нажал ctrl+alt+l
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split
import torch.optim as optim
from tqdm import tqdm # для прогресс бара
import os # нужно для проверки наличия модели
```

```python
train_df = pd.read_parquet('train.parquet', engine='fastparquet') # загрузка данных с fastparquet потому что parquet с чем то конфликтовал
test_df = pd.read_parquet('test.parquet', engine='fastparquet')
train_df['text'] = train_df['title'].fillna('') + ' ' + train_df['description'].fillna('') # предобработка чтобы с этим было проще работать
test_df['text'] = test_df['title'].fillna('') + ' ' + test_df['description'].fillna('')
train_df['price_log'] = np.log1p(train_df['price']) # логарифмирование цен потому чтобы они не так сильно отличались
tokenizer = AutoTokenizer.from_pretrained('DeepPavlov/rubert-base-cased') # токенизатор для того чтобы модель могла нормально работать с текстом
```

```python
class PriceModel(nn.Module):
    def __init__(self):
        super().__init__()
        print("загрузка модели")
        self.bert = AutoModel.from_pretrained(
            'DeepPavlov/rubert-base-cased')  # решил взять эту может потом возьму потяжелей
        self.fc = nn.Sequential( # делал несколько слоев
            nn.Linear(768, 256),   # сжатие
            nn.ReLU(),          # нелинейность
            nn.Dropout(0.3),       # от переобучения
            nn.Linear(256, 128), # снова сжатие
            nn.ReLU(),            # нелинейность
            nn.Dropout(0.2),  # вероятность уже поменьше 
            nn.Linear(128, 1) 
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids,
                            attention_mask=attention_mask)  # переход от того что думает модель к ответу
        last_hidden = outputs.last_hidden_state[:, 0, :]
        return self.fc(last_hidden).squeeze()
```

```python
class PriceDataset(Dataset):
    def __init__(self, df, tokenizer, is_train=True):  # инициализация
        self.df = df
        self.tokenizer = tokenizer
        self.is_train = is_train

    def __len__(self):  # тут все понятно
        return len(self.df)

    def __getitem__(self, idx):  # преобразование текста чтобы было понятно модели
        text = str(self.df.iloc[idx]['text'])
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',  # чтобы у всех было 352 контекста
            max_length=352, # это для модели с 352 контекста то есть 1 из ансамбля
            return_tensors='pt'
        )
        result = {
            'input_ids': encoding['input_ids'].squeeze(),
            # чтобы не добавлялась размерность батча а только то что нам надо
            'attention_mask': encoding['attention_mask'].squeeze()
            # добавлял padding поэтому надо чтобы модель понимала какие токены реальные
        }
        if self.is_train:
            result['price'] = torch.tensor(self.df.iloc[idx]['price_log'],
                                           dtype=torch.float32)  # метка дубликата
        return result
```

```python
train_data, val_data = train_test_split(train_df, test_size=0.1,
                                        random_state=42)  # 90% на обучение остальное на валидацию
train_dataset = PriceDataset(train_data, tokenizer, is_train=True)  # для токенизирования текста
val_dataset = PriceDataset(val_data, tokenizer, is_train=True)  # для обучения и валидации
test_dataset = PriceDataset(test_df, tokenizer, is_train=False)  # для тестирования
train_loader = DataLoader(train_dataset, batch_size=8, # pin memory понизил температуру видеокарты
                          shuffle=True, pin_memory=True)  # батч такой из-за ограничений vram, shuffle-перемешивание результатов
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)  # тут не принципиален батч сайз(т.к. нет градиентов)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)  # нет перемешивания результатов
```

```python
device = torch.device('cuda') # обучал на видеокарте
model = PriceModel().to(device)
if os.path.exists('best_model.pth'): # если есть то загружаюсь с существующей
    model.load_state_dict(torch.load('best_model.pth'))
optimizer = optim.Adam(model.parameters(), lr=5e-6)  # пока такой lr может изменю
criterion = nn.MSELoss()
best_val_loss = 9999999999 #изначально текое значение потом менял руками
for epoch in range(
        10):  # здесь не нужно столько эпох но я менял гиперпараметры руками и сам принимал решение когда мне остановится
    model.train()
    train_loss = 0
    for batch in tqdm(train_loader, desc= leave=False): # прогресс бар
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        prices = batch['price'].to(device)
        optimizer.zero_grad()  # обнуление градиентов
        preds = model(input_ids, attention_mask)  # forward
        loss = criterion(preds, prices)  # вычисление loss
        loss.backward()
        optimizer.step()  # обновление весов
        train_loss += loss.item()  # .item преобразует в число
    model.eval()
    val_loss = 0
    with torch.no_grad():  # валидация
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            prices = batch['price'].to(device)
            preds = model(input_ids, attention_mask)
            loss = criterion(preds, prices)
            val_loss += loss.item()
    avg_train_loss = train_loss / len(train_loader)  # вычисление ошибок за эпоху
    avg_val_loss = val_loss / len(val_loader)
    #вывод с точностью до 4 знаков
    print(f"эпоха {epoch + 1} train Loss: {avg_train_loss:.4f} val Loss: {avg_val_loss:.4f}")
    if avg_val_loss < best_val_loss:  # сохраняю на случай если захочу дальше обучать
        best_val_loss = avg_val_loss
        torch.save(model.state_dict(), 'best_model.pth')
        print("сохранена лучшая модель")
    else: # защита от переобучения
        break
```

```python
model.load_state_dict(torch.load('best_model.pth'))
model.eval()
predictions = []  # предсказания
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        preds = model(input_ids, attention_mask)
        predictions.extend(preds.cpu().numpy())
predictions = np.expm1(np.array(predictions))  # обратно из логарифма
results = pd.DataFrame({
    'id': test_df['id'].values,
    'y_pred': predictions
})
results.to_csv('submission.csv', index=False)  # сохранение результатов
print("все")
```
